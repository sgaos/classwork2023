```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
in_file = np.loadtxt("curve_to_match.dat", delimiter=',')
```

```python
time = in_file[0]
signal = in_file[1]
plt.plot(time, signal, "ro")
plt.xlabel("time")
plt.ylabel("signal")
plt.show()
```

```python
amplitude1 = 1
frequency1 = 2
frequency2 = 3
amplitude2 = 2
curve1 = amplitude1 * np.sin (frequency1 * time)
curve2 = amplitude2 * np.cos (frequency2 * time)
curve = curve1 + curve2
plt.plot(time, signal, "ro")
plt.plot(time, curve)
```
